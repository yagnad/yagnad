```python
import pandas as pd
import numpy as np
```

```python
#Read csv file using pandas
df = pd.read_csv("covid_cases.csv")
df.head()
```

```python
import plotly.express as px
# Confirmed
fig = px.choropleth(df, locations="Country_Region", 
                    locationmode='country names', color=np.log(df["Confirmed"]), 
                    hover_name="Country_Region", hover_data=['Confirmed'],
                    color_continuous_scale="Sunsetdark", 
                    title='Countries with Confirmed Cases')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```

```python
temp = df[df['Deaths']>0]
fig = px.choropleth(temp, 
                    locations="Country_Region", locationmode='country names',
                    color=np.log(temp["Deaths"]), hover_name="Country_Region", 
                    color_continuous_scale="rainbow", hover_data=['Deaths'],
                    title='Countries with Deaths Reported')
fig.update(layout_coloraxis_showscale=False)
fig.show()
```
